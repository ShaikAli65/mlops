### Analysis

1. How did GitHub Actions improve experiment reproducibility?

GitHub Actions ensured that every experiment was executed in a clean, consistent environment with the same dependencies and execution steps. This removed variability caused by local system differences and made each experiment fully reproducible.

2. How easy was it to compare results across runs?

Comparison was straightforward because each run produced standardized metrics (MSE and R^2), saved them as artifacts, and displayed them in the job summary. Different experiments could be compared directly by reviewing workflow runs.

3. What role does Git commit history play in experiment tracking?

Each commit represents a single experiment configuration. Commit messages document model type, preprocessing, and hyperparameters, creating a clear and traceable history of experimental changes.

4. Benefits compared to Lab 1

Compared to Lab 1's manual execution, this approach provided automation, reproducibility, and persistent experiment records. Results and models were automatically stored and could be revisited at any time.

5. Limitations of this approach

Hyperparameter tuning is manual and slow since each change requires a commit and workflow run. This approach also does not scale well for large hyperparameter searches compared to dedicated ML experiment tracking tools.
